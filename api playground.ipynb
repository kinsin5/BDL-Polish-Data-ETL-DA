```python
dataF = []
parent_id = "parent-id=G12"
response_G = requests.get(
    f'{baseAPI}{endPoint}?{lang_param}&{parent_id}'
).json() ## first call

while True:
    # Collect results from the current page
    dataF.append(pd.DataFrame(response_G['results']))
    # Check if there is a next page
    if "next" in response_G['links']:
        response_G = requests.get(
            f"{response_G['links']['next']}"
        ).json()
    else:
        break
    
finalDf = pd.concat(dataF, ignore_index=True)
finalDf
```

```python
## we're gathering data for all the years and all the vars all at once
page_size = "page-size=100"
endPoint = "data/by-variable/"
dfs = []

for _, row in subjectDf.iterrows():
    var_id = row['id']
    quarter = row['n1']
    sex = row['n2']

    try:
        response = requests.get(
            f"{baseAPI}{endPoint}{var_id}?{lang_param}&{page_size}"
            ).json()

        df = pd.DataFrame(response["results"])
        df["quarter"] = quarter
        df["sex"] = sex
        dfs.append(df)
    except Exception as e:
        print(f" Unexpected error for var_id={var_id}: {e}")
final_df = pd.concat(dfs, ignore_index=True)
```

```python
## we need to parse and exapand the data, to keep relevant column but to exapnd the values part
temp = final_df.explode('values', ignore_index=True)
temp['year'] = temp['values'].apply(lambda x: x['year'])
temp['val'] = temp['values'].apply(lambda x: x['val'])
df_expanded = temp.drop(columns=["values"])
```

```python
df_expanded["Year Quarter"] = (
    df_expanded["year"].astype(str)
    + " Q"
    + df_expanded["quarter"].str.split(" ").str[1]
)

# Extract numeric part of the quarter
df_expanded["quarter_num"] = df_expanded["quarter"].str.extract(r"(\d+)").astype(int)

# Map quarter number to month start
quarter_to_month = {1: 1, 2: 4, 3: 7, 4: 10}

# Compute the first month of the quarter
df_expanded["start_month"] = df_expanded["quarter_num"].map(quarter_to_month)

# Combine year and month into a date
df_expanded["Start of Quarter"] = pd.to_datetime(
    df_expanded["year"].astype(str) + "-" + df_expanded["start_month"].astype(str) + "-01"
)

# (Optional) drop helper columns
df_expanded.drop(columns=["quarter_num", "start_month"], inplace=True)
```
